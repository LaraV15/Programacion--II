### reindex()
Ahora supongamos que queremos reindexar al DF para asegurarnos que tiene los índices que nosotros queremos y también queremos que tenga todos los meses del año 2022

Para eso, vamos a usar el método reindex

`DataFrame.reindex(labels=None, index=None, columns=None, axis=None, method=None, copy=None, level=None, fill_value=nan, limit=None, tolerance=None)`

````python
# primero necesitamos crear el nuevo índice
# hay otras formas más rápidas para obtener estas combinaciones y les sugieron que las busquen en google
idx = []
for year in range(2023, 2013, -1):
  for month in range(12,1, -1): #Acá en vez de 13 va 12 porque hace referencia al mes
    idx.append('{}-{}-01'.format(year, month))
#idx
#el comando de arriba nos muestra todos los índices.

salario.reindex(idx)
````

Algo estamos haciendo mal porque no tiene sentido que todas las celdas sean nulas.
Qué pasa con los meses? Ven algo raro? 

````python
idx = []
for year in range(2023, 2013, -1):
  for month in range(13,1, -1):
    month_pad = str(month).rjust(2, "0")
    idx.append('{}-{}-01'.format(year, month_pad))

salario.reindex(idx)
````

````python
# también podemos hacer el reindex con alguna opción de llenado. 
salario.reindex(idx, fill_value = 'que mejoren')
# Y también puede hacerse la operación `inplace`
````

### Groupby
Nos sirve para generar grupos de datos usando una Serie del DF. En la [documentación](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.groupby.html) encontramos:
>A groupby operation involves some combination of splitting the object, applying a function, and combining the results. This can be used to group large amounts of data and compute operations on these groups.

````python
data = pd.read_csv('listings.csv.gz', compression = 'gzip') #no lo subi comprimido

#data = pd.read_csv('listings.csv')
````

````python
data.groupby('neighbourhood')
````

El resultado de groupby es un objeto DataFrameGroupBy. Generalmente no accedemos a este objeto porque no es necesario. Pero vamos a inspeccionarlo una vez para que sepan de que se trata.

````python
for name, gr in data.groupby('neighbourhood'):
  print(name)
  print(gr.columns)
  import pdb; pdb.set_trace()
````

````python
data['price_clean'] = data['price'].apply(lambda x: float(x.replace('$', '').replace(',', '')))
````

````python
# obtener el precio medio por barrio
data.groupby('neighbourhood')['price_clean'].mean()
````

````python
#agrupar por más de una columna
data.groupby(['neighbourhood', 'room_type'])['price_clean'].mean()
````

````python
#agrupar por más de una columna y calcularle la media a 2 columnas
data.groupby(['neighbourhood', 'room_type'])[['price_clean','minimum_nights']].mean()
````

````python
#Obtener la cantidad de registros para cada barrio y tipo de habitación
data.groupby(['neighbourhood', 'room_type']).size()
````

### Stack y Unstack
>At a high level, .unstack moves an index level into the columns. Usually we use this operation
on multi-index data, moving one of the indices into the columns (creating hierarchical columns).
The .stack method does the reverse, moving a multi-level column into the index
(Effective Pandas, M.Harrison)

````python
data.groupby(['neighbourhood', 'room_type'])['price_clean'].mean()
````

````python
data.groupby(['neighbourhood', 'room_type'])['price_clean'].mean().unstack()
````

````python
data[data['neighbourhood']=='Zwanenburg, Noord-Holland, Netherlands	']
````

````python
data.groupby(['neighbourhood', 'room_type'])['price_clean'].mean().unstack().stack()
````

### Merge
Para combinar 2 dataframes podemos usar pd.merge que nos permite hacerlo de forma similar a como se hace con las bases de datos. 
```
DataFrame.merge(right, how='inner', on=None, left_on=None, right_on=None, left_index=False, right_index=False, sort=False, suffixes=('_x', '_y'), copy=True, indicator=False, validate=None)
```
Vamos a usar los datos de:
- [Paso 2017](https://datos.gob.ar/dataset/interior-resultados-definitivos-elecciones-paso-2017)
- [Generales 2017](https://datos.gob.ar/dataset/interior-resultados-definitivos-elecciones-generales-2017)

````python
import pandas as pd
````

````python
paso = pd.read_csv('paso-2017definitivosndistrito.csv', sep = ';', encoding = 'latin-1')
paso.head()
````

````python
generales = pd.read_csv('generales-2017definitivosndistrito.csv', sep = ';', encoding = 'latin-1')
generales.head()
````

````python
#Unir dos DataFrame
merge = pd.merge(paso, generales, left_on=['DISTRITO','AGRUPACION POLITICA_TIPO DE VOTO'], 
                 right_on=['DISTRITO','AGRUPACION POLITICA_TIPO DE VOTO'], 
                 how = 'right')
merge.head()

#Queremos ver la tabla de las elecciones difinitivas y unirle una columna que corresponde
#A los votos que haya sacado en las PASO

#marge = pd.merge(nombre primera tabla, segunda tabla, del data set de la izquierda que concida
#con tales columnas del DataFrame de la derecha, y agregar a la derecha de la tabla la nueva columna,


#De todos modos en la ejecución tiene un problema cambiemos Bs. As tiene todos listas NaN y 
#cambiando juntos, para solcionarlo primero agrupalos (en el código que esta abajo del siguiente esta hecho)
````

````python
paso.groupby(['DISTRITO','AGRUPACION POLITICA_TIPO DE VOTO'])['VOTOS'].sum()
````

````python
paso['VOTOS'].fillna(0, inplace = True) #Reeplaza todos los nulos por true
generales['VOTOS'].fillna(0, inplace = True)
generales.loc[generales['VOTOS']=='   -  ', 'VOTOS'] = 0 # en una celda hay un - que nos molestas entonces lo sacamos
paso['VOTOS'] = paso['VOTOS'].apply(lambda x: int(str(x).replace('.', '')))
generales['VOTOS'] = generales['VOTOS'].apply(lambda x: int(str(x).replace('.', '')))
````

````python
paso_gr = paso.groupby(['DISTRITO','AGRUPACION POLITICA_TIPO DE VOTO'])['VOTOS'].sum()
paso_gr.head()
````

````python
merge = pd.merge(paso_gr, generales, left_index=True, 
                 right_on=['DISTRITO','AGRUPACION POLITICA_TIPO DE VOTO'], 
                 how = 'right', suffixes = ('_paso', '_generales'))
merge.head()
````

````python
merge[merge['DISTRITO']=='BUENOS AIRES'].sort_values('VOTOS_generales')
````

````python
merge[merge['DISTRITO']=='BUENOS AIRES'].set_index('AGRUPACION POLITICA_TIPO DE VOTO').VOTOS_paso.plot(kind = 'bar', color = 'b', alpha = 0.5)
merge[merge['DISTRITO']=='BUENOS AIRES'].set_index('AGRUPACION POLITICA_TIPO DE VOTO').VOTOS_generales.plot(kind = 'bar', color = 'r', alpha = 0.5)

#Diagrama de barras con los datos de las paso y de las elecciones generales
````
